```python
train_sample, test_sample = PrepareSubset(train_data, test_data, 1 )
```

# Step 3 - Extract features and enrich the dataset

```python
target_time_duration = 130
```

```python
def ExtractAudioFeatures(audio_array, sr=16000, n_mels=40, duration=2):

    audio = audio_array[:sr * duration]  # correct length =  (duration x sampling rate)
    mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels) # Compute Mel-spectrogram
    mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max) # Convert to dB scale (logarithmic scale)
    
    return mel_spectrogram.T  

def PadFeatures(features, target_length):
    # Pad sequences to the target length
    return np.pad(features, ((0, target_length - features.shape[0]), (0, 0)), mode='constant', constant_values=0)
```

```python
def PrepareDataWithFeatures(train_sample, target_time_duration):
    X_train = []
    y_train = []
    for sample in train_sample:
        audio_array = sample['audio']['array']
        label = sample['is_tts']
        
        # Extract features and append to training set
        mel_spec = ExtractAudioFeatures(audio_array)
        X_train.append(PadFeatures(mel_spec, target_time_duration))  # Pad features
        y_train.append(label)
    return X_train, y_train
```

```python
X_train = []
y_train = []

X_train, y_train = PrepareDataWithFeatures(train_sample, target_time_duration)
```

```python
X_test = []
y_test = []

X_test, y_test = PrepareDataWithFeatures(test_sample, target_time_duration)
```

```python
def ReshapeDataForCNN(X_train, y_train):
    X_train = np.array(X_train)
    y_train = np.array(y_train)
    X_train = X_train[..., np.newaxis]  # Adding a channel axis for CNN
    return X_train, y_train
```

```python
X_train, y_train = ReshapeDataForCNN(X_train, y_train)
X_test, y_test = ReshapeDataForCNN(X_test, y_test)                                  
```

# Step 4 - Now Define Model

```python
# Abstract out for experimentation
def DefineModel():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(X_train.shape[1], X_train.shape[2], 1)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid')  # Sigmoid for binary classification
    ])

    return model
```

```python
model = DefineModel()
```

# Step 5 - Prepare Model for training

```python
def SetTrainingArgs(model):
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = SetTrainingArgs(model)
```

# Step 6 - Start training

```python
def TrainModel(X_train, y_train, X_test, y_test, epochs=20):
    model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_data=(X_test, y_test))
    return model
```

```python
model = TrainModel(X_train, y_train, X_test, y_test, 20)
```

# Step 7 - Prepare submission

```python
def PrepareSubmission():
    y_pred_prob = model.predict(X_test)[:, 0] 
    df = pd.DataFrame({
        'id': [sample['id'] for sample in test_data], 
        'is_tts': y_pred_prob 
    })
    df.to_csv('./submission.csv', index=False)
    return df
```

```python
dfSubmission = PrepareSubmission()
```
